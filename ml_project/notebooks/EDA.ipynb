```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('heart.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
print(f'Количество дубликатов в данных: {data.duplicated().sum()}')
print(f'Количеcтво NAN значений в данных: {data.isna().sum().sum()}')
```

```python
categorical_features = ['sex', 'cp', 'restecg', 'exang', 'slope', 'ca', 'thal', 'fbs']
numerical_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(data[numerical_features].corr(), square=True, annot=True, ax=ax);
```

```python
fig, ax = plt.subplots(2, 4, figsize=(25, 10))
for variable, subplot in zip(categorical_features, ax.flatten()):
    sns.countplot(x=data[variable], hue=data['target'], ax=subplot, palette='YlGnBu')
```

```python
for variable in numerical_features:
    sns.displot(x=data[variable], hue=data['target'], kind='kde', fill=True, palette='YlGnBu')
```
